```python
import numpy as np
import matplotlib.pyplot as plt
```

$$\operatorname{arg min} \frac{1}{N} \sum_{i=1}^N l(x_i w, y)  + \lambda\|w\|_1$$

Logistic function

$$\frac{L}{1+ \exp{(-k(x-x_0))}}$$

```python
lambd = 0.1
l = 4*lambd
a = np.linspace(-l, l, 500).reshape(-1, 1)
b = np.linspace(-l, l, 500).reshape(1, -1)

def premise(a): # value == 0
    return np.maximum(0.0, 1 - a**2/(lambd**2))

def dpremise(a):
    return np.where(np.abs(a) < lambd, -2*a/lambd**2, 0.0)

def conclusion(a): # the value is not zero
    return a**2/lambd**2

def dconclusion(a):
    return 2*a/lambd**2
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
m = (1-a) * (1-b)
m = (1-a)**2 * (1-b)**2
m = premise(a) * conclusion(b)
ax[0].imshow(m, origin="lower", extent=(-l, l, -l, l))
ax[0].set_title("cost term")
m = premise(a) * dconclusion(b) + dpremise(a) * conclusion(b)
im = ax[1].imshow(-m, origin="lower", extent=(-l, l, -l, l), cmap="seismic")
ax[1].set_title("derivative")

cbar = fig.colorbar(im, ax=ax[1], location='right', anchor=(0, 0.3), shrink=0.7)

for i in range(2):
    ax[i].set_xlabel("premise")
    ax[i].set_ylabel("consequence")
```

```python
fig, ax = plt.subplots()
l1, = ax.plot(a, premise(a), label="$\\max(0, 1-\\frac{x^2}{\\lambda^2})$")
l2, = ax.plot(a, conclusion(a), label="$x^2$")
ax.plot(a, dpremise(a), ls=":", color=l1.get_color())
ax.plot(a, dconclusion(a), ls=":", color=l2.get_color())
l3, = ax.plot(a, dconclusion(a), ls=":", color=l2.get_color())
ax.set_ylim([-10, 10])
ax.legend()
```
